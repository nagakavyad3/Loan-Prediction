```python
data=pd.read_csv('loan_approval_dataset.csv')
```

```python
#Loan_id is not needed so that we can drop that column
data.drop(columns=['loan_id'],inplace=True)
```

```python
#For removing extra space 
data.columns=data.columns.str.strip()
```

```python
data.columns
```

```python
#Adding all assets into one
data['Assets'] = data.residential_assets_value + data.commercial_assets_value + data.luxury_assets_value + data.bank_asset_value
```

```python
data
```

```python
#Removing residential,commercial,luxury and bank assets
data.drop(columns = ['residential_assets_value','commercial_assets_value','luxury_assets_value','bank_asset_value'], inplace=True)
```

```python
#Checking duplicate values
data.isnull().sum()
```

```python
data
```

```python
#Converting string to numerical value
data.education.unique()
```

```python
#Removing extra space
def clean_data(st):
    st=st.strip()
    return st
```

```python
clean_data(' Graduate')
```

```python
data.education = data.education.apply(clean_data)
```

```python
data.education.unique()
```

```python
data['education'] = data['education'].replace(['Graduate', 'Not Graduate'],[1,0])
```

```python
data
```

```python
#For self_employed
data.self_employed.unique()
```

```python
data.self_employed = data.self_employed.apply(clean_data)
```

```python
data.self_employed.unique()
```

```python
data['self_employed'] = data['self_employed'].replace(['Yes', 'No'],[1,0])
```

```python
data
```

```python
#For loan_status
data.loan_status.unique()
```

```python
data.loan_status = data.loan_status.apply(clean_data)
```

```python
data.loan_status.unique()
```

```python
data['loan_status'] = data['loan_status'].replace(['Approved', 'Rejected'],[1,0])
```

```python
data
```

```python
#Splitting dataset into training and testing
from sklearn.model_selection import train_test_split
```

```python
input_data = data.drop(columns=['loan_status'])
output_data=data['loan_status']
```

```python
input_data
```

```python
output_data
```

```python
x_train,x_test,y_train,y_test =train_test_split(input_data,output_data,test_size=0.2)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
#Scaling the data with StandardScaler for improving the performance of the data
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
x_train_scaled = scaler.fit_transform(x_train)
```

```python
x_test_scaled = scaler.transform(x_test) #Scaling of input data is completed
```

```python
#Creating model we need logistic regression
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train_scaled, y_train) #Model is created
```

```python
#Evaluate model score with test dataset
model.score(x_test_scaled,y_test)
```

```python
#Lets predict some values
pred_data = pd.DataFrame([['2','1','0','9600000','2900000','12','778','50700000']], columns=['no_of_dependents',	'education', 'self_employed', 'income_annum', 'loan_amount', 'loan_term', 'cibil_score', 'Assets'])
```

```python
#Scale the above dataset
pred_data = scaler.transform(pred_data)
```

```python
#Predict the value
model.predict(pred_data) #Loan will be approved
```
